### Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

columns_to_drop = ['unhappy_flag1', 'unhappy_flag2', 'month_id', 'survey_date', 'subs_no', 'reason_for_score', 'survey_type', 'dom_cdmainid', 'main_score', 'tnps_segment']
data = data.drop(columns=[col for col in columns_to_drop if col in data.columns], errors='ignore')

data = data[data['brand'].str.lower() == 'xl']
```

```python
missing_data = data.isnull().sum()
missing_percentage = (missing_data / len(data)) * 100

missing_df = pd.DataFrame({
    'Kolom': missing_data.index,
    'Jumlah Kosong': missing_data.values,
    'Persentase Kosong (%)': missing_percentage.values
}).sort_values(by='Persentase Kosong (%)', ascending=False)
```

```python
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  
print(missing_df)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = data.select_dtypes(exclude=[np.number]).columns.tolist()

data[numeric_cols] = data[numeric_cols].fillna(0)
for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])  
# scaler = StandardScaler()
# data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

```

```python
from sklearn.preprocessing import  LabelEncoder

encoded_data = data.copy()

for col in categorical_cols:
    if data[col].nunique() <= 10:  
        one_hot = pd.get_dummies(data[col], prefix=col, drop_first=True)
        encoded_data = pd.concat([encoded_data, one_hot], axis=1)
    else:
        le = LabelEncoder()
        encoded_data[col] = le.fit_transform(data[col])

encoded_data = encoded_data.drop(columns=categorical_cols)
```

### Clustering

```python
inertia = []
cluster_range = range(1, 11)  

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(encoded_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o', linestyle='--')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inertia')
plt.title('Elbow Method untuk Menentukan Jumlah Cluster')
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=2, n_init=10)
data['Cluster'] = kmeans.fit_predict(encoded_data)
```

```python
data_unhappy = pd.read_csv('gilang_temp_gln_cx_chi_model_datamart_sample_v2.csv')[['unhappy_flag1', 'unhappy_flag2']]
data_unhappy = data_unhappy.loc[data.index]  

data['unhappy_flag1'] = data_unhappy['unhappy_flag1'].values
data['unhappy_flag2'] = data_unhappy['unhappy_flag2'].values

unhappy_props = data.groupby('Cluster')[['unhappy_flag1', 'unhappy_flag2']].mean() * 100
print("\nProporsi Unhappy Flag per Cluster:")
print(unhappy_props.to_string())


```

```python
categorical_cols = data.select_dtypes(exclude=[np.number]).columns.tolist()

valid_cat_cols = list(set(categorical_cols).intersection(data.columns))

if valid_cat_cols:
    cluster_modes = data.groupby("Cluster")[valid_cat_cols].agg(lambda x: x.mode().iat[0] if not x.mode().empty else None).T

    print("\nModus dari Kolom Kategorik per Cluster:")
    display(cluster_modes)
else:
    print("\nTidak ada kolom kategorik yang cocok ditemukan dalam dataset.")
```

```python
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()

valid_cols = list(set(numeric_cols).intersection(data.columns))

if valid_cols:
    cluster_means = data.groupby('Cluster')[valid_cols].mean()
    cluster_means = cluster_means.T
    

    cluster_means = cluster_means.sort_index()
    
    def highlight_max_min(s):
        max_val = s.max()
        min_val = s.min()
        return [
            "color: red; font-weight: bold;" if v == max_val else 
            "color: green; font-weight: bold;" if v == min_val else 
            "" 
            for v in s
        ]

    styled_df = cluster_means.style.apply(highlight_max_min, axis=1).format("{:,.2f}")

    display(styled_df)
else:
    print("\nTidak ada kolom numerik yang valid ditemukan dalam dataset.")
```

```python
cluster_counts = data['Cluster'].value_counts().sort_index()

cluster_counts_df = pd.DataFrame({
    'Cluster': cluster_counts.index,
    'Frekuensi': cluster_counts.values
})


cluster_counts_df.style.format({"Frekuensi": "{:,.0f}"})
```

```python
data_rec = pd.read_csv('gilang_temp_gln_cx_chi_model_datamart_sample_v2.csv')[['main_score', 'tnps_segment']]
data_rec = data_rec.loc[data.index]  

data['main_score'] = data_rec['main_score'].values
data['tnps_segment'] = data_rec['tnps_segment'].values

def categorize_score(score):
    if 9 <= score <= 10:
        return '10-9'
    elif 7 <= score < 9:
        return '8-7'
    elif 0 <= score < 7:
        return '6-0'
    else:
        return None

data['main_score_category'] = data['main_score'].apply(categorize_score)

result = data.groupby(['Cluster', 'main_score_category']).size().reset_index(name='Count')

print(result)

# result.to_csv('main_score_cluster_summary.csv', index=False)
```
